## This notebook shows a short visualization of some variables in the data that contributed to World Happiness Report 2018. I decided to use a subset of the dataset that focuses on countries in Africa in order to put  things in perspective.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Happiness = pd.read_csv('WorldHappiness2018_Data.csv')
```

```python
#A short view of the first 5 records in the dataset
Happiness.head()
```

```python
#All the variables that were considered in calculing the happiness score
Happiness.columns
```

```python
#Total number of countries in the dataset
len(Happiness.Country)
```

```python
Happiness.isna().sum()
```

```python
#ALl the names of countries in the dataset
Happiness.Country.unique()
```

```python
Africa = ['Libya', 'Algeria', 'Morocco', 'Nigeria', 'Somalia', 'Cameroon', 'Gabon', 'South Africa', 'Ivory Coast', 'Ghana',
          'Senegal', 'Tunisia', 'Sierra Leone', 'Congo (Brazzaville)', 'Mali', 'Namibia','Burkina Faso', 'Egypt',
          'Mozambique', 'Kenya', 'Zambia', 'Mauritania', 'Ethiopia', 'Chad', 'Congo (Kinshasa)','Niger', 'Uganda', 'Benin', 'Sudan',
          'Togo', 'Guinea', 'Lesotho', 'Angola', 'Madagascar', 'Zimbabwe', 'Botswana', 'Malawi', 'Liberia', 'Rwanda',  
          'Tanzania', 'South Sudan', 'Central African Republic', 'Mauritius', 'Burundi']
```

```python
#An extraction of all the records for African countries in the original data
Africa_data = Happiness.loc[Happiness['Country'].isin(Africa)]
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=80)
sns_plot = sns.barplot(x="Country", y="Rank", data = Africa_data)
sns.set_context('notebook', font_scale = 1.5)
plt.title("Happiness rank of countries in Africa")
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(16, 8))
plt.xticks(rotation=80)
sns_plot = sns.barplot(x="Country", y="Generosity", data = Africa_data)
sns.set_context('notebook', font_scale = 1.5)
plt.title("The level of generosity of different countries in Africa")
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=80)
sns.set_context('notebook', font_scale = 1.5)
sns_plot = sns.barplot(x="Country", y="Healthy_Life_Expectancy", data = Africa_data)
plt.title("Healthy life expectancy of countries in Africa")
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=80)
sns.set_context('notebook', font_scale = 1.5)
sns_plot = sns.barplot(x="Country", y="Freedom_To_Make_Life_Choices", data = Africa_data)
plt.title("The level of freedom to make life choices in Africa")
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=80)
sns.set_context('notebook', font_scale = 1.5)
sns_plot = sns.barplot(x="Country", y="Social_Support", data = Africa_data)
plt.title("The level at which countries in Africa get social support")
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=80)
sns.set_context('notebook', font_scale = 1.5)
sns_plot = sns.barplot(x="Country", y="GDP_Per_Capita", data = Africa_data)
plt.title("The Gross Domestic Product Per Capital of Countries in Africa")
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize=(15, 8))
plt.xticks(rotation=80)
sns.set_context('notebook', font_scale = 1.5)
sns_plot = sns.barplot(x="Country", y="Perceptions_Of_Corruption", data = Africa_data)
plt.title("The level at which different countries in Africa perceive corruption")
```

## #The jointplots below show a positive relationship between happiness score and the selected variables. 

```python
sns.set(style = 'whitegrid')
sns.set_context('notebook', font_scale = 1.5)
sns_plot = sns.jointplot(x="GDP_Per_Capita", y="Score", data = Africa_data, height = 6)
```
